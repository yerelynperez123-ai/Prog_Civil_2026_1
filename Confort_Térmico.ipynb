## Teoría del Confort Térmico en Ganado Vacuno

El **confort térmico** en el ganado vacuno se refiere a un estado en el cual el animal no experimenta estrés por calor ni por frío, lo que le permite mantener su temperatura corporal central dentro de rangos fisiológicos normales con un mínimo esfuerzo metabólico. Este estado es crucial para la salud, bienestar y productividad de los animales, ya que las desviaciones del confort térmico pueden tener impactos negativos significativos.

### Factores que Influyen en el Confort Térmico:

1.  **Temperatura Ambiental (Temperatura de Bulbo Seco - Tbs):** Es el factor más obvio. Temperaturas muy altas o muy bajas pueden causar estrés.
2.  **Humedad Relativa (Temperatura de Bulbo Húmedo - Tbh):** La humedad afecta la capacidad del animal para disipar calor a través de la evaporación (sudoración y respiración). Alta humedad en combinación con altas temperaturas exacerba el estrés por calor.
3.  **Radiación Solar:** La exposición directa al sol aumenta la carga de calor en el animal.
4.  **Velocidad del Viento:** El viento puede ayudar a la disipación de calor por convección en ambientes cálidos o aumentar la pérdida de calor en ambientes fríos.
5.  **Precipitación:** La lluvia puede enfriar al animal, pero también puede mojar el pelaje, reduciendo su capacidad aislante en climas fríos.
6.  **Características del Animal:** Edad, raza, estado fisiológico (lactancia, gestación), nivel de producción y color del pelaje influyen en su tolerancia al estrés térmico.

### Estrés por Calor y Estrés por Frío:

*   **Estrés por Calor:** Ocurre cuando la carga de calor total sobre el animal excede su capacidad para disipar ese calor. Esto lleva a un aumento de la temperatura corporal, cambios hormonales, reducción del consumo de alimento, disminución de la producción de leche y carne, y problemas reproductivos. El ganado vacuno es particularmente susceptible al estrés por calor debido a su rumia, que genera calor metabólico.
*   **Estrés por Frío:** Sucede cuando la temperatura ambiental cae por debajo de la zona termoneutral del animal, obligándolo a aumentar su producción de calor metabólico para mantener la temperatura corporal. Esto puede llevar a un mayor consumo de alimento, pérdida de peso, y disminución de la productividad.

### Importancia del Índice de Temperatura y Humedad (ITH):

El ITH es una medida utilizada comúnmente para evaluar el nivel de estrés térmico al que está expuesto el ganado. Combina la temperatura del aire y la humedad relativa en un solo valor, proporcionando una indicación más precisa del ambiente que experimenta el animal en comparación con la temperatura sola. Los umbrales de ITH permiten clasificar el nivel de estrés (no estresante, alerta, peligro, crítico) y sirven como una herramienta valiosa para que los productores ganaderos implementen estrategias de mitigación y adaptación, como proporcionar sombra, acceso a agua fresca, ventilación, o modificar los horarios de alimentación.

# **Métodos y cálculos**

+ Índice de temperatura y humedad (producción intensiva)

$$ITH = 0.72 * (T_{bs} + T_{bh}) + 40.6$$

Donde:

1. $T_{bs}$ Temperatura bulbo seco ($^oC$)
2. $T_{bh}$ Temperatura de bulbo húmedo ($^oC$)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda utilizar los límites propuestos por National Weather Service:

1. Sí $ITH <= 70$ ambiente no estresante
2. Sí $ITH > 70 \space and \space ITH <= 78$ situación de alerta por estrés térmico
3. Sí $ITH > 78 o <= 83$ peligro por estrés térmico
4. Sí $ITH > 83$ situación crítica


````python
#ENTRADA: pedir tbs (temperatura de Bulbo seco) y tbh (temperatura Bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, Grados Centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, Grados Centigrados: "))

#PROCESO: calculo de ith (indice de temperatura y humedad)
ith = 0.72 * (tbs + tbh) + 40.6

#SALIDA: imprimir resultados de la evalucion del ith y el confort térmico
if ith <= 70:
  print("el ambiente es no estresante")
elif ith > 70 and ith <= 78:
  print("la situacion es de alerta por estrés térmico")
elif ith > 78 and ith <= 83:
  print("peligro por estrés térmico")
else:
  print("situacion critica")
````
