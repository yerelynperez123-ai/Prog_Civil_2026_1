# ✅ Paso 1: Planteamiento del problema de ingeniería (sencillo)

Trabajemos con un problema básico aplicado a **Ingeniería Civil**:

> Calcular el **volumen de concreto** necesario para una zapata rectangular.

En ingeniería, el volumen es:

[
V = largo \times ancho \times altura
]

---

# ✅ Paso 2: Definir Entradas, Proceso y Salidas

Este es el corazón del pensamiento algorítmico.

## 🔹 1. Entradas (Datos que el usuario debe proporcionar)

* Largo de la zapata (m)
* Ancho de la zapata (m)
* Altura de la zapata (m)

## 🔹 2. Proceso (Cálculo que realiza el algoritmo)

* Multiplicar largo × ancho × altura

## 🔹 3. Salida (Resultado que muestra el programa)

* Volumen de concreto en metros cúbicos (m³)

---

# ✅ Paso 3: Diseño del Algoritmo (en palabras)

1. Pedir al usuario el largo.
2. Pedir el ancho.
3. Pedir la altura.
4. Calcular el volumen multiplicando los tres valores.
5. Mostrar el resultado.

---

# ✅ Paso 4: Pseudocódigo

```
INICIO

    Leer largo
    Leer ancho
    Leer altura

    volumen ← largo * ancho * altura

    Mostrar volumen

FIN
```

Observa que aquí estamos definiendo claramente variables y usando el operador de multiplicación `*`.

---

# ✅ Paso 5: Código en Python (comentado y explicado)

```python
# Programa para calcular el volumen de una zapata rectangular
# Autor: Estudiante de Ingeniería
# Descripción: Calcula el volumen de concreto en m³

# Entrada de datos
largo = float(input("Ingrese el largo de la zapata (m): "))
ancho = float(input("Ingrese el ancho de la zapata (m): "))
altura = float(input("Ingrese la altura de la zapata (m): "))

# Proceso (cálculo)
volumen = largo * ancho * altura

# Salida
print("\nEl volumen de concreto necesario es:", volumen, "m³")
```
