```markdown
## SESIÓN 1: Introducción a los Procesos Iterativos

**¿Por qué necesitamos ciclos?**
Imagina que debes calcular el promedio de temperatura durante los 7 días de la semana. Sin usar ciclos, tendrías que escribir líneas de código para pedir la temperatura siete veces distintas. Un ciclo (o bucle) nos permite escribir esa instrucción una sola vez y decirle a Python que la repita automáticamente.

### 1. El Bucle `while` (Mientras que...)
La estructura `while` ejecuta un bloque de código continuamente *mientras* una condición siga siendo verdadera.

Para evitar que el ciclo se repita infinitamente, utilizamos un **contador**: una variable que aumenta su valor en cada iteración para llevar el control.

```

### Celda 3: Código (Python)

````
# Ejemplo de bucle while controlado por un contador
contador_dias = 1

# El ciclo se ejecutará mientras el contador sea menor o igual a 5
while contador_dias <= 10:
    print("Registrando datos para el día número:", contador_dias, "cuadrado:", contador_dias**2)
    # Aumentamos el contador en 1 para que el ciclo avance
    contador_dias = contador_dias + 2
````

```markdown
### 2. El Bucle `for` (Para...)
El bucle `for` es ideal cuando sabemos exactamente *cuántas veces* necesitamos repetir una acción. En Python, solemos combinarlo con la función `range()`, la cual genera una secuencia de números.

```

### Celda 5: Código (Python)

````
list(range(1,6))
````

````
# Ejemplo de bucle for
# range(5) genera automáticamente 5 valores: 0, 1, 2, 3, 4
b = 2
for h in range(1,10,2):
    print("Area igual para h = ", h, "El area es = ", b*h/2)
````

```markdown
## SESIÓN 2: Aplicación Práctica con Acumuladores

En ingeniería no solo contamos eventos, también sumamos cantidades a lo largo del tiempo. Para esto usamos un **acumulador**: una variable que suma o "acumula" valores progresivamente durante el ciclo.

### Mini-proyecto 1: Promedio de Mediciones Térmicas
Vamos a automatizar la recolección de 5 mediciones de temperatura en un proceso agroindustrial y calcularemos su promedio.

```

### Celda 8: Código (Python)

````
# Programa para promediar 5 mediciones de temperatura
suma_temperaturas = 0  # Inicializamos el acumulador en cero

for i in range(5):
    # Se repite 5 veces la petición de datos al usuario
    temp_ingresada = float(input("Ingrese la temperatura medida (°C): "))

    # Acumulamos el valor ingresado a la suma total
    suma_temperaturas = suma_temperaturas + temp_ingresada

# Al terminar el ciclo, calculamos y mostramos el promedio
promedio = suma_temperaturas / 5
print("\nLa temperatura promedio del proceso es:", promedio, "°C")
````

```markdown
### Mini-proyecto 2: Registro Hidrológico
Este programa simula el registro de la precipitación diaria durante una semana en una estación meteorológica local, calculando el total de lluvia acumulada, un dato vital en el diseño de drenajes agrícolas o civiles.

```

### Celda 10: Código (Python)

````
# Programa de registro de precipitación semanal
lluvia_total = 0  # Nuestro acumulador

for dia in range(7):
    # Usamos (dia + 1) para que en pantalla se vea "día 1" en lugar de "día 0"
    lluvia_diaria = float(input(f"Ingrese la lluvia registrada el día {dia + 1} (en mm): "))
    lluvia_total = lluvia_total + lluvia_diaria

print("\nLa precipitación total acumulada en la semana es de:", lluvia_total, "mm")
````
